```python
import numpy  as np
import pandas  as pd
import nltk
import re
import os
from sklearn import feature_extraction
import pickle
import scipy
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
from nltk import word_tokenize
from nltk.corpus import stopwords
```

```python
titles = pickle.load(open('data/titles.pkl','rb'))
genres = pickle.load(open('data/genres.pkl','rb'))
synopsis = pickle.load(open('data/synopses.pkl','rb'))
print(len(titles), len(genres), len(synopsis))
```

```python
print(synopsis[0][:1000])
```

```python
def preprocess_data(text):
    
    tagged_sentence = nltk.tag.pos_tag(text.split())
    text = [word for word,tag in tagged_sentence if tag != 'NNP' and tag != 'NNPS']
    text=' '.join(text)  
    
    result = nltk.RegexpTokenizer(r'\w+').tokenize(text)
    text=' '.join(result)
    stop = set(stopwords.words('english'))
    s=[]
    for i in text.split():
        if i not in stop:
            s.append(i)
            
    text=' '.join(s) 
    text=text.lower()
    m=[]
    for t in text.split():
        k=wnl.lemmatize(t, pos='v')
        m.append(k)
   
    text=' '.join(m)

    return text
```

```python
 preprocess_data(synopsis[0][:1000])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(max_df=0.8, max_features=20000, min_df=0.2,\
                                   stop_words='english', preprocessor=preprocess_data, \
                                   use_idf=True, ngram_range=(1, 1))

%time tfidf_matrix = tfidf_vectorizer.fit_transform(synopsis)

print(tfidf_matrix.shape)
terms = tfidf_vectorizer.get_feature_names()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(tfidf_matrix.todense())
```

```python
import pandas as pd
films = {'title': titles, 'synopsis': synopsis, 'cluster': kmeans.labels_, 'genre': genres}
frame = pd.DataFrame(films, index = [kmeans.labels_] , columns = ['title', 'cluster', 'genre'])
frame
```

```python
frame['cluster'].value_counts()
```

```python
print("Top terms per cluster:")
order_centroids = np.asarray(kmeans.cluster_centers_).argsort()[:, ::-1]
cluster_names = []
for i in range(order_centroids.shape[0]):
    print("Cluster %d words:" % i, end='')
    q = ""
    for ind in order_centroids[i, :6]:
        print(' %s' % terms[ind], end=',')
        q += str(terms[ind])
        q += " "
    cluster_names.append(q)
    print()
    print("Cluster %d titles:" % i, end='')
    for title in frame.loc[i]['title'].values.tolist():
        print(' %s,' % title, end='')
    print()
 
```

```python
import matplotlib.pyplot  as plt
from sklearn.decomposition import PCA as sklearnPCA
```

```python
pca = sklearnPCA(n_components=2) #2-dimensional PCA
transformed = pd.DataFrame(pca.fit_transform(tfidf_matrix.todense()))
```

```python
colors = ['red', 'blue', 'green', 'yellow', 'black', 'gray', 'orange', 'brown']
for i in range(len(cluster_names)):
    plt.scatter(transformed[kmeans.labels_ == i][0], transformed[kmeans.labels_ == i][1], label=cluster_names[i], c=colors[i])

plt.show()
```
